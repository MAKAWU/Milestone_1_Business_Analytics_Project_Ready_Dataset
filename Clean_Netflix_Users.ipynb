```python
import pandas as pd
```

```python
# Load the dataset
file_path = r"C:\Users\Maryam\Desktop\BAN6800_Business_Analytics_Capstone\Kaggle_Netflix_Users_Database_Ext\Uncleaned_Netflix_Users.csv"

df = pd.read_csv(file_path)
```

```python
# 1. Convert Last_Login to datetime format
df['Last_Login'] = pd.to_datetime(df['Last_Login'], errors='coerce')

# 2. Drop duplicate rows
df_cleaned = df.drop_duplicates()

# 3. Remove rows with invalid Age or Watch_Time_Hours (e.g., negative values)
df_cleaned = df_cleaned[(df_cleaned['Age'] > 0) & (df_cleaned['Watch_Time_Hours'] >= 0)]

# 4. Standardize text columns to lowercase
text_columns = ['Name', 'Country', 'Subscription_Type', 'Favorite_Genre']
for col in text_columns:
    df_cleaned[col] = df_cleaned[col].str.lower()

# 5. Reset index after cleaning
df_cleaned.reset_index(drop=True, inplace=True)

# Display cleaned dataset
df_cleaned.head()
```
